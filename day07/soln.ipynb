```python
l={'shiny gold'}
d={}
for run in range(22):
    for line in data.split('\n'):
        a,b=line.split('contain')
        a=a.split(' ')
        out_colour=a[0]+' '+a[1]
        if out_colour in l:
            l.remove(out_colour)
            for contain in b.split(','):
                if 'no other' in contain:
                    continue
                else:
                    contain=contain.split(' ')
                    in_colour=contain[2]+' '+contain[3]
                    num=int(contain[1])
                    l.add(in_colour)
                    if out_colour in d:
                        if (num,in_colour) not in d[out_colour]:
                            d[out_colour].append((num,in_colour))
                    else:
                        d[out_colour] = [(num, in_colour),]
```

```python
def getTotal(col):
    if col not in d:
        return 1
    else:
        total=0
        for i in d[col]:
            if i[1] not in d:
                total+=i[0]
            else:
                total+=i[0]
                total+=(i[0]*getTotal(i[1]))
        return total
```
